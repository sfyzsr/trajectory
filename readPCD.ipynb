```python
import numpy as np 
from pypcd import pypcd
pc = pypcd.PointCloud.from_path("3/trajectory.pcd")
pc_data = pc.pc_data
pc_array = np.array([pc_data["x"], pc_data["y"], pc_data["z"]], dtype=np.float32)
pc_array = np.transpose(pc_array)
pc_array.shape
```

```python
import matplotlib.pyplot as plt

x = pc_array[:, 0]  
y = pc_array[:, 1]  

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('(X, Y) Point Cloud')
plt.show()
pc_array
```

```python
np.save("3.npy",pc_array)
```
